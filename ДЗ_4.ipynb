**Задача 1. Случайная непрерывная величина A имеет равномерное распределение на
промежутке (200, 800].
Найдите ее среднее значение и дисперсию.**

Воспользуемся следующими формулами:  

$M(X)=\frac{a+b}{2}$ 

$D(X)=\frac{(b-a)^2}{12}$



```python
print (f'M = {(200+800)/2}')
print (f'D = {(800-200)**2/12}')
```
